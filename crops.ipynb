```python
import os
from PIL import Image, ImageDraw
import numpy as np
import matplotlib.pyplot as plt
import skimage
import numpy as np
from skimage import measure
from skimage import io, color, img_as_ubyte
from skimage.feature import greycomatrix, greycoprops
from sklearn.metrics.cluster import entropy
```

```python
dr = '/Users/hpathak/Downloads/Urban100_SR/image_SRF_4/'
paths =  os.listdir(dr)

hr_paths = [_ for _ in paths if _.endswith('HR.png')]
hr_paths = sorted(hr_paths)
sr_paths = [_ for _ in paths if _.endswith('ExSR.png')]
sr_paths = sorted(sr_paths)
image_list_HR = [os.path.join(dr, _) for _ in hr_paths]
image_list_SR = [os.path.join(dr, _) for _ in sr_paths]

#measure.compare_psnr(HR,SR)
```

```python

lr_paths = [_ for _ in paths if _.endswith('LR.png')]
lr_paths = sorted(lr_paths)
image_list_LR = [os.path.join(dr, _) for _ in lr_paths]
```

```python

psnr_list = []
for i,j in zip(image_list_HR,image_list_SR):
    i = skimage.io.imread(i)
    j = skimage.io.imread(j)
    psnr_list.append(measure.compare_psnr(i,j))
print(psnr_list)
```

```python
sobel_list = []
for j in image_list_LR:
    #print(np.gradient(np.asarray(j)))
    j = io.imread(j)
    grayImg = img_as_ubyte(color.rgb2gray(j))
    sobel_list.append(sum(sum(skimage.filters.sobel(grayImg))))
sobel_list  = [(_/max(sobel_list))**2 for _ in sobel_list]
```
